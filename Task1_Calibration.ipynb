Standard imports for data analysis and plotting.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read in the calibration data - skipping over the info at the head of the CSV file.

```python
calib = pd.read_csv("calibration.csv", header=10)
```

Separate the two samples (this makes the indexing a lot simpler!). We can also drop the columns we're not interested in.

```python
S1 = calib.loc[calib.Sample == 'S1'].drop(columns=['Well','Sample'])
Blank = calib.loc[calib.Sample == 'Blank'].drop(columns=['Well','Sample','Dilution'])
```

To plot, calculate the mean (y-value) and standard error ($\frac{\sigma}{\sqrt{N}}$) (y-error) for each triplicate measurement.

```python
S1_mean = S1.groupby('Dilution').mean()
S1_err = S1.groupby('Dilution').std()/np.sqrt(3)
B_mean = Blank.mean()
B_err = Blank.std()/np.sqrt(3)
x = S1.columns.values[1:].astype('int')
```

Plot the data, with errors.

```python
plt.figure(figsize=(10,6))
for i in range(len(S1_mean)):
    plt.plot(x, S1_mean.iloc[i], label=S1_mean.iloc[i].name)
    plt.fill_between(x, (S1_mean-S1_err).iloc[i], (S1_mean+S1_err).iloc[i], alpha=0.5)
plt.plot(x, B_mean, label='Blank')
plt.fill_between(x, (B_mean-B_err), (B_mean+B_err), alpha=0.5)
plt.xlabel('$\lambda$ (nm)')
plt.ylabel('$_{log}$(Absorbance)')
plt.yscale('log')
plt.legend(title='Dilution')
plt.title("Absorbance (Raw) vs Wavelength")
plt.show()
```

Given that the absorbance of a sample $x$ is $A_x = \sum_i A_{x_i}$ for all constituent compounds $x_i$, to correct for the solvent, simply subtract the absorbance of the blank sample.

```python
plt.figure(figsize=(10,6))
for i in range(len(S1_mean)):
    plt.plot(x, (S1_mean-B_mean).iloc[i], label=S1_mean.iloc[i].name)
    plt.fill_between(x, ((S1_mean-B_mean)-(S1_err+B_err)).iloc[i], ((S1_mean-B_mean)+(S1_err+B_err)).iloc[i], alpha=0.5)
plt.xlabel('$\lambda$ (nm)')
plt.ylabel('$_{log}$(Absorbance)')
plt.yscale('log')
plt.legend(title='Dilution')
plt.title("Absorbance (Corrected) vs Wavelength")
plt.show()
```
